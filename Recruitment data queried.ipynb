```python
##What is the average age of applicants by education level?
```

```python
df["Date of Birth"] = df["Date of Birth"].str.replace("/", "-", regex=False)
df["Date of Birth"] = pd.to_datetime(df["Date of Birth"], format="%d-%m-%Y", dayfirst=True)
```

```python
df["Age"] = (datetime.now() - df["Date of Birth"]).dt.days // 365
```

```python
df.groupby('Education Level')['Age'].mean()
```

```python
##Which cities have the highest number of applicants?
```

```python
df['City'].value_counts()
```

```python
##What is the gender distribution for each job title category?
```

```python
Gender_count_by_Job_title = df.groupby("Job Title")["Gender"].value_counts().reset_index( name = "Gender_count")
Gender_count_by_Job_title 
```

```python
##What is the average desired salary by years of experience?
```

```python
df.groupby("Years of Experience")["Desired Salary"].mean().reset_index()
```

```python
##How does desired salary vary by education level?
```

```python
df.groupby("Education Level")["Desired Salary"].mean().reset_index()
```

```python
##Which job titles have the highest average desired salary?
```

```python
df.groupby("Job Title")["Desired Salary"].mean().reset_index().sort_values(by = "Desired Salary", ascending = False ).head(5)
```

```python
##Which months had the highest number of applications?
```

```python
df.groupby("Application_month")["Applicant ID"].count().idxmax()
```

```python
"""July has the highest number of applications"""
```

```python
###What percentage of applicants in each education level received offers?
```

```python
total_applicants = df.groupby("Education Level")["Education Level"].count()
offered = df[df['Status'] == "Offered"].groupby("Education Level")["Applicant ID"].count()
offer_percent = (offered/total_applicants) * 100
offer_percent
```

```python
###Is there a correlation between years of experience and desired salary?
```

```python
df['Years of Experience'].corr(df['Desired Salary'])	
```

```python
##Which job titles have the highest rejection rates?
```

```python
df[df["Status"] == "Rejected"].groupby("Job Title")["Applicant ID"].count().idxmax()
```

```python
##What is the acceptance rate by country of origin?
```

```python
df[df["Status"] == "Offered"].groupby("Country")['Applicant ID'].count().reset_index()
```

```python
##How many Chief Technology Officer applicants are there and what are their average qualifications?
```

```python
chief_tech_officer = df[df["Job Title"] == "Chief Technology Officer"]
chief_tech_officer.groupby("Education Level")['Years of Experience'].mean().reset_index()
```
